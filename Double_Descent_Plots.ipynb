```
def process_list(l):
  train_errors = []
  test_errors = []
  aucs = []
  for idx, i in enumerate(l):
    if idx % 3 == 0:
      train_errors.append(i)
    elif idx % 3 == 1:
      test_errors.append(i)
    else:
      aucs.append(i)
  return train_errors, test_errors, aucs
```

```
def plot(data, title='', log=False, gen_error=False):
  train_errors, test_errors, aucs = process_list(data)
  xs = list(range(1, len(aucs)+1))
  plt.figure(figsize=(5,3))
  if log:
    plt.xscale('log')

  if gen_error:
    plt.plot(xs, [i-j for i, j in zip(test_errors, train_errors)], label='Generalization Error (Te - Tr)')
  else:
    plt.plot(xs, train_errors, label='Train Errors')
    plt.plot(xs, test_errors, label='Test Errors')
  plt.plot(xs, aucs, label='AUCs')
  plt.legend(loc='best')
  plt.title(title)
  plt.xlabel('Epochs')
  plt.ylabel('Error / AUC')
  plt.show()
```
